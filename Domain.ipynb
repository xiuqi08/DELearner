```python
import pandas as pd
import numpy as np
import os
import re
```

```python
dic = {"a":[269,1879,2288,2311,2747,2790,6707,7442,7521,7557,7423],
        "b":[614,2840,7432],
        "c":[653,840,1389,5404,7660,7682,7689],
        "d":[2006,2765,3308,3362,4250,5402,6421],
        "e":[175,634,553,2333,2480,3317,4496],
        "f":[2074,4365,5356,6439,7504,7720,7721,7763],
        "g":[818,4398,4358,4316,4315,5297,5371,5399],
        "h":[1541,1997,2005,2963,3661],
        "i":[204,2352,2702]}
```

```python
lines = []
try:
    for line in open(os.path.join("./","Reference.txt")):
        if (len(line) > 10):
            lines.append(line)
except:
    pass
```

```python
def splitNoAuthor(line):
    s = re.split(r"([0-9]+)",line)
    no = s[1]
    author = s[2].replace("\n","")
    return no,author
line = "269Blei DM, Ng AY, Jordan MI"
no,author = splitNoAuthor(line)
print("line",line)
print("no",no)
print("author",author)
```

```python
df = pd.DataFrame()
for ind in range(len(lines)-2):
    if lines[ind][0] in [str(n) for n in list(range(10))]:
        line = lines[ind]
        no,author = splitNoAuthor(line)
        title = lines[ind+1].replace("\n","")
        ref = lines[ind+2].replace("\n","")
        data = pd.DataFrame([no,author,title,ref]).T
        data.columns=["No","Author","Title","Reference"]
        df = pd.concat([df,data],axis=0)
        ind += 2
ref = df.reset_index(drop=True)
print(ref.shape)
ref.head()
```

```python
cols = ["Theme","No","Author","Title","Reference","Initiate"]
df = pd.DataFrame(columns=cols)
cnt = 0
for theme,items in dic.items():
    for no in items:
        cnt += 1
        data = []
        data.append(theme)
        data.append(str(no))
        if ref.loc[ref['No']==str(no)].shape[0] > 0:
            data.append(ref.loc[ref['No']==str(no),"Author"].values[0])
            data.append(ref.loc[ref['No']==str(no),"Title"].values[0])
            data.append(ref.loc[ref['No']==str(no),"Reference"].values[0])
            if no == items[0]:
                data.append(1)
            else:
                data.append(0)
            data = pd.DataFrame(np.array(data).T).T
            data.columns = cols
            df = pd.concat([df,data],axis=0)
print(cnt)
df = df.reset_index(drop=True)
df
```
